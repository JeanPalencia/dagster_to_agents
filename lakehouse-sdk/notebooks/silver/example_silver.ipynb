## 3. Probar Joins entre Silver Assets


```python
# Load multiple assets silver
df_appointments = execute_asset("stg_s2p_appointments", "2024-12-01")
df_projects = execute_asset("stg_s2p_projects", "2024-12-01")

# Experiment with join
df_joined = df_appointments.join(
    df_projects,
    left_on="project_id",
    right_on="project_id",
    how="inner"
)

print(f"Appointments: {df_appointments.height}, Projects: {df_projects.height}, Joined: {df_joined.height}")
df_joined.head()
```
