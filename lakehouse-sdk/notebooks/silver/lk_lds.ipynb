```python
df = execute_asset("core_project_funnel", partition_key="2024-12-01")
# Agregar campos constantes de auditoría
import polars as pl

aud_inserted_at_str = "2025-12-15 08:17:34.726521"
aud_inserted_date_str = "2025-12-15"
aud_updated_at_str = "2025-12-15 08:17:55.183409"
aud_updated_date_str = "2025-12-15"
aud_job_str = "lk_projects"

# La versión moderna de polars ha cambiado el argumento a "date_format" y "time_format"
# Pero para .str.strptime(), simplemente usa "format" como argumento 
# Referencia: https://pola-rs.github.io/polars/py-polars/html/reference/expressions/api/polars.Expr.str.strptime.html

df = df.with_columns([
    pl.lit(aud_inserted_at_str).str.strptime(pl.Datetime("ms"), format="%Y-%m-%d %H:%M:%S%.f").alias("aud_inserted_at"),
    pl.lit(aud_inserted_date_str).str.strptime(pl.Date, format="%Y-%m-%d").alias("aud_inserted_date"),
    pl.lit(aud_updated_at_str).str.strptime(pl.Datetime("ms"), format="%Y-%m-%d %H:%M:%S%.f").alias("aud_updated_at"),
    pl.lit(aud_updated_date_str).str.strptime(pl.Date, format="%Y-%m-%d").alias("aud_updated_date"),
    pl.lit(aud_job_str).alias("aud_job"),
])

# Guardar el DataFrame como archivo CSV
# Exportar el DataFrame a CSV sin convertir nulls en cadenas vacías
# En polars, por default, los nulls quedan como '', lo que da problema con PostgreSQL
# Solución: serializar a csv usando null_value="\\N" (el default de psql/copy), o usar otra convención que no sea cadena vacía ni comillas

#df.write_csv("lpf_lead_project_funnel.csv")
df
```

```python
# Revisa si hay registros duplicados por project_id (usando Polars)
duplicate_projects = (
    df_f
    .group_by("project_id")
    .agg(pl.count().alias("count"))
    .filter(pl.col("count") > 1)
)
duplicate_projects
```

```python
# Crea un DataFrame con los registros duplicados y agrupados por project_id
duplicate_project_ids = (
    df_f
    .group_by("project_id")
    .agg(pl.count().alias("count"))
    .filter(pl.col("count") > 1)
    .select("project_id")
)

df_f_duplicates = df_f.join(duplicate_project_ids, on="project_id", how="inner")

df_f_duplicates
```
