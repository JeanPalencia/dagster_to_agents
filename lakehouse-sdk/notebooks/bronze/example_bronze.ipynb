# Lakehouse SDK - Bronze Layer Testing

This notebook demonstrates how to use the Lakehouse SDK to test Bronze layer assets.

## Setup


```python
from lakehouse_sdk import (
    clear_cache,
    load_all_assets,
    execute_asset,
    list_assets,
    quick_profile,
    validate_asset,
    ValidationRule,
    print_validation_results,
    CommonRules,
)

import polars as pl

clear_cache()
```

## 1. Discover Available Assets


```python
# List all available assets
list_assets()
```

## 2. Execute Bronze Asset: raw_s2p_clients


```python
# Execute asset with a specific partition
df_clients = execute_asset("raw_s2p_clients", partition_key="2024-12-01")

# Quick profile
quick_profile(df_clients, "Raw S2P Clients")
```

```python
# Execute asset with a specific partition
df_clients = execute_asset("raw_s2p_project_requirements", partition_key="2024-12-01")

# Quick profile
quick_profile(df_clients, "Raw S2P Projects Requirements")
```

```python
df_clients.head(10)
```

```python
# Limpiar el cache
clear_cache()

# Ahora ejecutar el asset
df = execute_asset('raw_s2p_project_requirements', '2024-12-01')

# Verificar
print(f"Tipo de reason_id: {df['reason_id'].dtype}")
print(f"Valores únicos: {df['reason_id'].unique().drop_nulls().sort()}")
```

```python
df = execute_asset('raw_s2p_project_requirements', '2024-12-01')

# Ver distribución de reason_id
print("Distribución de reason_id:")
print(df['reason_id'].value_counts().sort("reason_id"))

# Ver total de nulls
null_count = df['reason_id'].null_count()
print(f"\nTotal nulls: {null_count} de {df.height} ({null_count/df.height*100:.1f}%)")

# Ver si hay otros valores
print(f"\nValores únicos en reason_id: {df['reason_id'].unique().drop_nulls().sort()}")
```

## 3. Explore Data


```python
# View first rows of specific columns
df_clients.select(['id', 'email', 'client_type', 'client_state', 'created_at']).head(10)
```

```python
# Client type distribution
df_clients.group_by('client_type').agg(pl.len().alias('count')).sort('count', descending=True)
```

## 4. Validate Data


```python
# Define validation rules
rules = [
    ValidationRule("id", CommonRules.not_null(), "ID must not be null"),
    ValidationRule("id", CommonRules.unique(), "ID must be unique"),
    ValidationRule("email", CommonRules.not_null(), "Email must not be null"),
]

# Execute validations
results = validate_asset(df_clients, rules)
print_validation_results(results)
```

## 5. Test Joins (Composite Transformations)

Before creating a new asset, you can test joins and transformations here.


```python
# Example: Load another asset and join
df_profiles = execute_asset("raw_s2p_profiles", partition_key="2024-12-01")

df_joined = df_clients.join(
    df_profiles,
    left_on="id",
    right_on="client_id",
    how="left"
)

quick_profile(df_joined, "Clients + Profiles")
```
